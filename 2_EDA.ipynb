```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo de gráficos
sns.set(style='whitegrid')

# Cargar datos LIMPIOS
df = pd.read_csv('./data/titanic_clean.csv')

df.head()
df.info()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Distribución de la variable objetivo (Survived)')
plt.xlabel('Survived (0 = no, 1 = sí)')
plt.ylabel('Conteo')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Supervivencia por sexo')
plt.xlabel('Sexo')
plt.ylabel('Conteo')
plt.legend(title='Survived', labels=['No', 'Sí'])
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Supervivencia por clase (Pclass)')
plt.xlabel('Clase')
plt.ylabel('Conteo')
plt.legend(title='Survived', labels=['No', 'Sí'])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Distribución de la edad (Age)')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Distribución de la tarifa (Fare) según supervivencia')
plt.xlabel('Survived (0 = no, 1 = sí)')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.heatmap(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(), 
            annot=True, 
            cbar=True,  # Activar la barra de colores
            cmap='coolwarm',  # Cambiar la paleta de colores
            cbar_kws={'label': 'Nivel de correlación'})  # Etiqueta para la barra de colores
plt.title('Correlación de variables numéricas')
plt.show()
```

```python
# Cálculo de la tasa de supervivencia
survival_rate = df['Survived'].mean() * 100
non_survival_rate = 100 - survival_rate

print(f"Tasa de supervivencia: {survival_rate:.2f}%")
print(f"Tasa de no supervivencia: {non_survival_rate:.2f}%")
```
